```python
import PyPDF2


def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            text += pdf_reader.pages[page_num].extractText()
    return text

file_path = "/home/gcouser/Scrivania/bimby.pdf"
extracted_text = extract_text_from_pdf(file_path)
print(extracted_text)

with open("extracted-text.txt", 'w') as file:
    file.write(extracted_text)

```

```python
# QUESRTA HA FUNZIONATO MEGLIO
import pdfplumber

def extract_text_from_pdf(file_path):
    with pdfplumber.open(file_path) as pdf:
        text = ' '
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Specifica il percorso del tuo file PDF
file_path = "/home/gcouser/Scrivania/bimby.pdf"

# Estrai il testo dal PDF
extracted_text = extract_text_from_pdf(file_path)

# Stampa il testo estratto
print(extracted_text)
with open("v2.txt", 'w') as file:
    file.write(extracted_text)
    
```
